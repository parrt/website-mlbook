```python
class MyVec:
    def __init__(self, values):
        self.data = values
    def __add__(self, other):
        newdata = [x+y for x,y in zip(self.data,other.data)]
        return MyVec(newdata)
    def __str__(self):
        return '['+', '.join([str(v) for v in self.data])+']'

a = MyVec([1,2,3])
b = MyVec([3,4,5])
print(a + b)
print(a.__add__(b)) # how a+b is implemented
```
